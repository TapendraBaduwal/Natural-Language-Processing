```python
from bs4 import BeautifulSoup
import requests
title_list =[]
for i in range(16):
  #print("page", i)
  url = requests.get(f'https://merojob.com/job-level/entry_level/?page={i}')
  soup = BeautifulSoup(url.text, 'html.parser')
  story1 = soup.find_all("h1", class_="text-primary font-weight-bold media-heading h4")
  
  for q in story1:
      title = q.text
      title_list.append(title)


print(title_list)
      
```

```python
from bs4 import BeautifulSoup
import requests
company_list = []
for i in range(16):
  url = requests.get(f'https://merojob.com/job-level/entry_level/?page={i}')
  soup = BeautifulSoup(url.text, 'html.parser')
  story1 = soup.find_all("h3", class_="h6")
    
  for q in story1:
      company = q.text
      company_list.append(company)
      #print(company)   
print(company_list)
      
```

```python
import pandas as pd
dataset = pd.DataFrame({"Job_Title": title_list,"Company_Name":company_list})
dataset
```
